# Parameter estimation techniques in Lotka-Volterra model

In this document, four parameter estimation techniques are presented (efforts will be made to compare them), applied on a simple **Lotka-Volterra model** for two populations, i.e. with one population acting as a predator of the other. First, we will do a _linear regression_ on the parameters. Following this, we will modify the linear regression using _tikhonov regularization_. Next, we do a so-called _nonlinear least squares regression_. These first three techniques, being amply employed, don't require much further presentation. That is not the case with the other two. We will follow the treatment found in Ramsay and Hooker (2017); they refer to the techniques as _gradient matching_ and _parameter cascading_.

First of all, we define some necessary functions.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import linalg
from scipy import integrate
from scipy import optimize
from scipy import interpolate
from sklearn.linear_model import Ridge
plt.style.use('ggplot')

def lv_equation(t, x, *args):
    """
    Returns the Lotka-Volterra differential equation for two populations.
    Should receive six parameters (r1, r2, b11, b12, b21, b22)
    """
    r1 = args[0]
    b12 = args[1]
    r2 = args[2]
    b21 = args[3]
    r = np.array([r1, r2])
    b = np.array([[0, b12], [b21, 0]])
    return x * (r + b @ x)  # x_i = x_i (r_i + \sum b_ij x_j)

def add_noise(data, mean=0, sigma=1):
    """
    Implements additive Gaussian noise with mean 0 and standard deviation sigma
    to each entry of the given data
    """
    return data + np.random.normal(loc=mean, scale=sigma, size=np.shape(data))

def mult_noise(data, mean=1, sigma=1):
    """
    Implements multipliactive Gaussian noise with mean 1 and standard deviation sigma
    to each entry of the given data
    """
    return data * np.random.normal(loc=mean, scale=sigma, size=np.shape(data))

def clean_data(data, eps):
    """
    Reduces to zero the entries of data with value < eps
    """
    data[data < eps] = 0
    return data
```

## Data generation

First, the necessary parameters are defined. Here, $x_0$ is an array containing the initial population sizes, while $t_0$ and $t_f$ define the interval of the ODE to be integrated. Finally, we generate $t_{points}$, an array containing the time points in which the solution of the ODE is to be evaluated.

Next, we generate noisy data. For this purpose, we first generate data with the previous parameters and add Gaussian noise with mean $0$ and standard deviation $\sigma$. Then, we simply set the values below a certain threshold $\epsilon$ to zero. We show plots of the original data and the noisy data side to side.

```python
# ODE Parameters
r1 = 1
r2 = -1.5
b12 = -0.3
b21 = 0.7
params = (r1, b12, r2, b21)

# Additional conditions
t0 = 0     
tf = 15
x0 = np.array([3,10])

# Time points to be displayed
num_tpoints = 100
t_points = np.linspace(t0, tf, num_tpoints)

# Noise parameters
sigma = 5/np.sqrt(num_tpoints)

# Data below this value is considered to be zero
eps = 1E-4

# Data generation and stuff
sol_lv = integrate.solve_ivp(lv_equation, (t0,tf), x0, args=params, t_eval=t_points)
data_lv = add_noise(sol_lv.y, 0, sigma)
data_lv = clean_data(data_lv, eps)

# Plotting
fig, (ax0, ax1) = plt.subplots(1, 2, figsize=(12,5))
ax0.plot(t_points, sol_lv.y[0], '-r', label='true x0')
ax0.plot(t_points, sol_lv.y[1], '-b', label='true x1')
ax0.set_title('Clean data')
ax0.set_ylabel('population')
ax0.set_xlabel('time')
ax0.legend(loc='upper left')
ax1.plot(t_points, data_lv[0], 'r*', label='exp x0')
ax1.plot(t_points, data_lv[1], 'b*', label='exp x1') 
ax1.set_title('Noisy data')
ax1.set_xlabel('time')
ax1.legend(loc='upper left')
plt.show()
```

## Parameter estimation

In the following treatment of linear regression with and without tikhonov regularization, we follow the treatment in Stein et al. (2013).

### Linear least squares

We follow the treatment in Stein et al. (2013). First we define the appropriate matrices and vectors so that we have an equation of the form $A x = b$, arranged so that $b$ is matrix of 'differentials' (roughly, the left side of the original differential equation), while x is a matrix containing all the ODE parameters.

\begin{align}
&A_1 x_1 = b_1 \\
&A_2 x_2 = b_2
\end{align}

```python
b1 = np.zeros(num_tpoints)
b2 = np.zeros(num_tpoints)
A1 = np.zeros((num_tpoints, 2))
A2 = np.zeros((num_tpoints, 2))
A1[:,0] = np.ones(num_tpoints)    # sets first column equal to ones
A2[:,0] = np.ones(num_tpoints)    # ditto
for i in range(num_tpoints-1):
    if data_lv[0,i+1]==0 or data_lv[0,i]==0:
        b1[i] = 0
    else:
        b1[i] = np.log(data_lv[0,i+1]/data_lv[0,i])/(t_points[i+1]-t_points[i])
    if data_lv[1,i+1]==0 or data_lv[1,i]==0:
        b2[i] = 0
    else:
        b2[i] = np.log(data_lv[1,i+1]/data_lv[1,i])/(t_points[i+1]-t_points[i])
    A1[i,1] = data_lv[1,i]
    A2[i,1] = data_lv[0,i]

reg1 = optimize.lsq_linear(A1, b1)    # returns (r1, b12)
reg2 = optimize.lsq_linear(A2, b2)    # returns (r2, b21)

ls_fitted_params = tuple(reg1.x) + tuple(reg2.x)

print("The original parameters were:", params)
print("The fitted parameters are: ", ls_fitted_params)
print(np.abs(np.array(ls_fitted_params) - np.array(params)))

ls_fitted_data = integrate.solve_ivp(lv_equation, (t0,tf), x0, args=ls_fitted_params, t_eval=t_points)

plt.plot(t_points, ls_fitted_data.y[0], 'r-', label='fit')
plt.plot(t_points, ls_fitted_data.y[1], 'b-', label='fit')
plt.plot(t_points, data_lv[0], 'm*', label='data')
plt.plot(t_points, data_lv[1], 'g*', label='data')
plt.legend(loc='best')
plt.xlabel('Time')
plt.ylabel('Population')
plt.show()
```

#### A "manual" way to write the solution to $A_j x = b_j$

```python
x1 = linalg.inv(np.transpose(A1) @ A1) @ np.transpose(A1) @ b1
x2 = linalg.inv(np.transpose(A2) @ A2) @ np.transpose(A2) @ b2
new_params = tuple(x1) + tuple(x2)
print(params)
print(new_params)
print(np.abs(np.array(params)-np.array(new_params)))
```

## Regularization procedures
There is often bias in the solution to linear least squares problem. Overfitting and stuff. Solution: regularization. New problem: regularization parameter must be chosen, so what criteria should we use to choose it?

Our objective is to minimize the function $\lVert Gm-d \rVert^2_2 + \alpha^2 \lVert m \rVert^2_2$. Our strategy will be to create a new matrix, so that we end up with the usual least squares problem.

$$ \min \Bigg\lVert \begin{bmatrix} G \\ \alpha \mathbf{I} \end{bmatrix} m 
- \begin{bmatrix} d \\ 0 \end{bmatrix}\Bigg\rVert^2_2 $$

This is the same as solving for $G^T G + \alpha^2 \mathbf{I}) m = G^T d$.

But how do we choose which is the best $\alpha$? The first idea that we present is using the so-called *L-curve criterion*.

### L-curve criterion

This method consists in trying for different values of $\alpha$ and graphing y vs x. One must consider many orders of magnitude.

Here is an implementation:

```python
n_alpha = 200
alpha = np.logspace(-10,10,n_alpha)
reg_params_list = np.zeros((n_alpha, 4))
#reg_fit_data = np.zeros((n_alpha, 2, num_tpoints))
reg_params_norms = np.zeros(n_alpha)
reg_fit_data_norms = np.zeros(n_alpha)
for i in range(n_alpha):
    reg_params1 = linalg.inv( (np.transpose(A1) @ A1) + (alpha[i]**2*np.diag(np.ones(2))) ) @ np.transpose(A1) @ b1
    reg_params2 = linalg.inv( (np.transpose(A2) @ A2) + (alpha[i]**2*np.diag(np.ones(2))) ) @ np.transpose(A2) @ b2
    reg_params = tuple(reg_params1) + tuple(reg_params2)
    #reg_params_list[i] = np.array(reg_params)
    reg_fit = integrate.solve_ivp(lv_equation, (t0,tf), x0, args=reg_params, t_eval=t_points)
    #reg_fit_data[i] = reg_fit.y
    reg_params_list[i] = np.array(reg_params)
    reg_params_norms[i] = linalg.norm(reg_params_list[i])
    reg_fit_data_norms[i] = linalg.norm(data_lv - reg_fit.y)

#print(reg_params_norms)
#print(reg_fit_data_norms)
plt.plot(reg_fit_data_norms, reg_params_norms, 'o');
plt.ylabel('norm of solution')
plt.xlabel('norm of error')
#plt.yscale('log')
#plt.xscale('log')
plt.show()
```

## Nonlinear Least Squares

We will use a hybrid method: Our parameter estimates obtained in the ordinary least squares procedure will be used as initial estimates for a nonlinear least squares optimization.

```python
def lv_residuals(params):
    sim_data = integrate.solve_ivp(lv_equation, (t0,tf), x0, args=params, t_eval=t_points)
    return sim_data.y[0] - data_lv[0] + sim_data.y[1] - data_lv[1]

nls_fitted_params = optimize.least_squares(lv_residuals, ls_fitted_params)

initial_guess = integrate.solve_ivp(lv_equation, (t0,tf), x0, args=ls_fitted_params, t_eval=t_points)
nls_fitted_data = integrate.solve_ivp(lv_equation, (t0,tf), x0, args=nls_fitted_params.x, t_eval=t_points)

#plt.plot(t_points, initial_guess.y[0], 'b-', label='Initial Guess')
#plt.plot(t_points, initial_guess.y[1], 'b-', label='Initial Guess')
plt.plot(t_points, nls_fitted_data.y[0], 'r-', label='Fitted Parameters')
plt.plot(t_points, nls_fitted_data.y[1], 'b-', label='Fitted Parameters')
plt.plot(t_points, data_lv[0], 'm*', label='Data')
plt.plot(t_points, data_lv[1], 'g*', label='Data')
plt.legend(loc='best')
plt.xlabel('Time')
plt.ylabel('Population')
plt.show()
```
